```sql
-- Query 7: The "Root Cause" Analysis
-- Group failures by reason. Is it Schema Mismatch? Nulls?
SELECT _failed_cols, count(*) as failure_volume 
FROM quarantine_audit 
GROUP BY _failed_cols 
ORDER BY failure_volume DESC;

-- Query 8: The "Recent Incidents" Log
-- Show me the errors from the last 24 hours only.
SELECT date, symbol, _failed_cols, ingestion_timestamp 
FROM quarantine_audit 
ORDER BY ingestion_timestamp DESC 
LIMIT 20;

-- Query 9: "False Positive" Check
-- Are we rejecting good data? Check a specific rejected row to see if the logic was too strict.
SELECT * FROM quarantine_audit 
WHERE _failed_cols LIKE '%eps%' -- Focus on EPS failures
LIMIT 5;
```
