# Performance Optimization: Multi-Dimensional Clustering (Z-Order)

## Objective
This notebook executes the `OPTIMIZE` command with `ZORDER BY` on critical Delta tables. 

## Z-Order Strategy
Z-Ordering co-locates related information in the same set of files. This technique dramatically reduces the amount of data the engine needs to read (Data Skipping) when queries filter by high-cardinality columns like `symbol`.

*   **Target Column:** `symbol`
*   **Reasoning:** Most downstream analytical queries and dashboards filter or join based on the Ticker Symbol. Co-locating this data minimizes I/O overhead.

```python
%sql
OPTIMIZE gold_valid_audit ZORDER BY (symbol)
```

```python
%sql
OPTIMIZE "/Volumes/workspace/default/storage/gold/ticker_data_v6" ZORDER BY (symbol)
```

```python
%sql
OPTIMIZE "/Volumes/workspace/default/storage/serving/valuation_dashboard_v1" ZORDER BY (symbol)
```
