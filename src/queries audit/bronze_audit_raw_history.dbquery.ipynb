```sql
-- Query 1: The "Rescue" Check
-- Did we catch any malformed JSON that would have crashed a lesser pipeline?
SELECT count(*) as rescued_count, _rescued_data 
FROM bronze_audit 
WHERE _rescued_data IS NOT NULL 
GROUP BY _rescued_data;

-- Query 2: The "Duplicates" Proof
-- We expect duplicates here because we ran "Surgical Backfills". 
-- This proves Bronze is a true Append-Only history.
SELECT symbol, date, count(*) as versions_count 
FROM bronze_audit 
GROUP BY symbol, date 
HAVING count(*) > 1 
ORDER BY versions_count DESC;

-- Query 3: Schema Evolution Verification
-- Do we have the new 'ingestion_timestamp' and correctly typed 'eps'?
SELECT symbol, date, ingestion_timestamp, eps, typeof(eps) as eps_type
FROM bronze_audit
ORDER BY ingestion_timestamp DESC 
LIMIT 5;
```
