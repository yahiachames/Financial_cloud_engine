```sql
-- Query 4: The "Idempotency" Certificate
-- CRITICAL: This MUST return 0 rows. 
-- If it returns rows, your Window Function/Deduplication logic failed.
SELECT symbol, date, count(*) 
FROM silver_audit 
GROUP BY symbol, date 
HAVING count(*) > 1;

-- Query 5: Data Quality (Type Enforcement)
-- Verify 'eps' is a Decimal (money) and not null for major companies.
SELECT symbol, date, eps, revenue 
FROM silver_audit 
WHERE eps IS NULL OR revenue IS NULL
AND symbol IN ('AAPL', 'MSFT', 'NVDA'); -- Major tickers shouldn't be null

-- Query 6: Partition Health
-- Ensure we haven't accidentally deleted history during a "Surgical Backfill".
SELECT date, count(*) as records_per_day 
FROM silver_audit 
GROUP BY date 
ORDER BY date DESC;
```
