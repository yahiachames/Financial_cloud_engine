# Ad-Hoc Analysis: Beta & Volatility Statistics

## Purpose
This query performs statistical analysis on the distribution of Beta (Market Risk) values across the dataset to identify outliers and central tendencies.

```sql
SELECT 
    s AS ticker,
    COUNT(*) AS total_rows,
    
    -- Range Check
    ROUND(MIN(beta), 3) AS min_beta,
    ROUND(MAX(beta), 3) AS max_beta,
    
    -- Central Tendency
    ROUND(AVG(beta), 3) AS avg_beta,
    ROUND(percentile_approx(beta, 0.5), 3) AS median_beta, -- This is the 50th percentile
    
    -- Real Volatility (Standard Deviation)
    ROUND(STDDEV(beta), 3) AS beta_std_dev
    
FROM delta.`/Volumes/workspace/default/storage/gold/ticker_data_v10`
GROUP BY s
ORDER BY avg_beta DESC;



SELECT 
    symbol AS ticker,
    COUNT(*) AS total_rows,
    
    -- Range Check
    ROUND(MIN(beta), 3) AS min_beta,
    ROUND(MAX(beta), 3) AS max_beta,
    
    -- Central Tendency
    ROUND(AVG(beta), 3) AS avg_beta,
    ROUND(percentile_approx(beta, 0.5), 3) AS median_beta, -- This is the 50th percentile
    
    -- Real Volatility (Standard Deviation)
    ROUND(STDDEV(beta), 3) AS beta_std_dev
    
FROM delta.`/Volumes/workspace/default/storage/serving/valuation_dashboard_v5`
GROUP BY symbol
ORDER BY avg_beta DESC;



```
