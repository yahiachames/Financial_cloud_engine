# View Persistence Configuration
This script establishes the **Serving Layer Logic** by creating persistent views over the underlying Delta storage in Unity Catalog Volumes.

```python
%sql
DROP TABLE IF EXISTS serving_layer_ticker_stream;
```

```python
%sql
CREATE OR REPLACE VIEW gold_layer_ticker_stream AS 
SELECT * FROM delta.`/Volumes/workspace/default/storage/gold/ticker_data_v10`;
```

```python
%sql
CREATE OR REPLACE VIEW serving_layer_ticker_stream AS 
SELECT * FROM delta.`/Volumes/workspace/default/storage/serving/valuation_dashboard_v5`;
```
